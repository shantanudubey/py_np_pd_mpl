```python
# DataFrame from Dictionary : where the keys are the column labels and the values are the column values
df = pd.DataFrame(
    {
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo"
    }
)
df
```

```python
# Have to use at least one vector if using dictionary 
df = pd.DataFrame(
    {
        "key_01": 1,
        "key_02": 2,
        "key_03": "value_03",
        "key_04": [1, 2, 3]
    }
)
df
```

```python
df.describe()
```

```python
# Sort by index
df.sort_index(axis=1, ascending=False)
```
