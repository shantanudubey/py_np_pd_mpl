# Matplotlib

```python
"""
    Matplotlib : https://matplotlib.org/stable/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py
        - pyplot is a collection of functions that make matplotlib work like MATLAB.
        - all sequences are converted to numpy arrays internally. 
        
    This starts with the FSDS class and adds some playing around with the pyplot tutorial.
"""

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
x
```

```python
y
```

```python
# Size of x and y has to be same else we get an error
plt.scatter(x,y)
```

```python
# Change Parameters : Color can also use rgb codes like #e377c2
plt.scatter(x,y,c='y')
plt.title("Graph in 2D")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.savefig("files/test_mplb.png")
```

```python
y = x * x
plt.plot(x,y,'#1177cf')
```

```python
# Reperesentation options : 
# color : like r* b* y* r+ b+ bo go etc.
plt.plot(x,y,'m--')
```

```python
# style options : 
plt.plot(x,y,'g*', linestyle="dashed", linewidth=3, markersize=18)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("2D Graph")
```

```python
# Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--',linewidth=3)

plt.subplot(2,2,2)
plt.plot(y,x,'g*', markersize=9)

plt.subplot(2,2,3)
plt.plot(y,x,'b+', markersize=9)

plt.subplot(2,2,4)
plt.plot(x,y,'mo',markersize=9)
```

```python
# Bar plots : 

x1 = [2,3,4,5]
y1 = [11,16,19,20]
x2 = [3,9,11,14]
y2 = [6,15,7,9]

plt.bar(x1,y1)
plt.bar(x2,y2,color='g')
```

```python
# Histograms : 
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27,99,100,100,101,200,200,199,198,195, 145,146,147,149,150])
plt.hist(a)
```

```python
# Pie Chart :
labels = "python", "c++", "ruby", "java"
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = [0.4,0,0,0]

plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=False, autopct='%1.1f%%')
plt.axis("equal")
plt.show()
```

```python
data = {
        'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)
        }
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# Plot using categorical variables

users = ["user_01", "user_02", "user_03", "user_04"]
values = [1, 10, 50, 100]

plt.figure(figsize=(12,3))

plt.subplot(231)
plt.bar(users, values)

plt.subplot(232)
plt.scatter(users, values)

plt.subplot(233)
plt.plot(users, values)

plt.subplot(234)
plt.plot(users, values, "g--")

plt.subplot(235)
plt.plot(values, users, "r--")

plt.suptitle("Categorical Plotting")
plt.show()
```
